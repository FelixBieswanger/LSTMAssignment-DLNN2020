```python
import matplotlib.pyplot as plt
import json
import numpy as np
```

```python
data_rnn = dict()
with open("losstracking/elman-rnn_loss.json", "r") as file:
    data_rnn = json.loads(file.read())


data_lstm = dict()
with open("losstracking/lstm_loss.json", "r") as file:
    data_lstm = json.loads(file.read())
```

```python

fig,ax = plt.subplots(figsize=(20,10))
plt.plot(list(data_rnn.keys()),list(data_rnn.values()))
plt.plot(list(data_rnn.keys()), list(data_lstm.values()))
plt.ylabel("Loss")
plt.xlabel("iterations")
ax.set_xticks(ax.get_xticks()[::100])
plt.legend(["rnn", 'lstm'], loc='upper right')
plt.savefig("loss_comparison.png")

```
